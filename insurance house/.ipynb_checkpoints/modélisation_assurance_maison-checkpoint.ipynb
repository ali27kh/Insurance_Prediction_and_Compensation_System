## Calcule des colonnes data 1

### 1)Age:

```python
def calculate_age(dt1):
    dob_column_name = 'date_of_birth'
    dt1['date_of_birth'] = pd.to_datetime(dt1['date_of_birth'])
    dt1['Age'] = dt1[dob_column_name].apply(lambda x: 2019 - x.year if x.year < 2019 else 0)
    return dt1
```

```python
calculate_age(dt1)
```

### 2)Properties_Market_Value:

```python
dt1['Properties_Market_Value'] = dt1['home_market_value'] / 10 + 1000
```

```python
dt1
```

```python
dt1.isna().sum()
```

```python
def calculate_first_contract_duration(dt1):
    dob_column_name1 = 'Date_first_contract'
    dt1['Date_first_contract'] = pd.to_datetime(dt1['Date_first_contract'])
    dt1['Duration'] = dt1[dob_column_name1].apply(lambda x: 2019 - x.year if x.year < 2019 else 0)
    return dt1
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df= dt[['CITY', 'marital_status']]
```

```python
def encode_df(dataframe):
    le = LabelEncoder()
    for column in dataframe.columns:
        dataframe[column] = le.fit_transform(dataframe[column])
    return dataframe

#encode the dataframe
df
```

```python
# RÃ©organiser l'ordre des colonnes
dt = dt[['Client_id', 'Duration', 'Age', 'has_children', 'home_market_value', 'Properties_Market_Value',
       'Vulnerable to theft', 'home_owner', 'Annual_prime',
       'Insured_Amount', 'Nombre_Thefts_last5years', 'Potential client']]
```

```python
dt = pd.concat([dt, df], axis=1)
```

```python
dt
```

```python
features = ['CITY', 'Duration', 'Age', 'has_children',
       'marital_status', 'home_market_value', 'Properties_Market_Value',
       'Vulnerable to theft', 'home_owner', 'Annual_prime',
       'Insured_Amount', 'Nombre_Thefts_last5years']
label = 'Potential client'
```

```python
X, y = dt[features].values, dt[label].values
```

```python
y[2]
```

```python
# Split data 70%-30% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
print ('Training cases: %d\nTest cases: %d' % (X_train.shape[0], X_test.shape[0]))
```

```python
# Train the model
from sklearn.linear_model import LogisticRegression

# Set regularization rate
reg = 0.01

# train a logistic regression model on the training set
model = LogisticRegression(C=1/reg, solver="liblinear").fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print('Predicted labels: ', predictions)
print('Actual labels:    ', y_test)
print('Accuracy: ', accuracy_score(y_test, predictions))
```

```python
from sklearn. metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import precision_score

print("Overall Precision:", precision_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix

# Print the confusion matrix
cm = confusion_matrix(y_test, predictions)
print (cm)
```
