## Calcule des colonnes data 1

### 1)Age:

```python
def calculate_age(dt1):
    dob_column_name = 'date_of_birth'
    dt1['date_of_birth'] = pd.to_datetime(dt1['date_of_birth'])
    dt1['Age'] = dt1[dob_column_name].apply(lambda x: 2019 - x.year if x.year < 2019 else 0)
    return dt1
```

```python
calculate_age(dt1)
```

### 2)Properties_Market_Value:

```python
dt1['Properties_Market_Value'] = dt1['home_market_value'] / 10 + 1000
```

```python
def calculate_first_contract_duration(dt1):
    dob_column_name1 = 'Date_first_contract'
    dt1['Date_first_contract'] = pd.to_datetime(dt1['Date_first_contract'])
    dt1['Duration'] = dt1[dob_column_name1].apply(lambda x: 2019 - x.year if x.year < 2019 else 0)
    return dt1
```

```python
columns_to_extract_dt1 = ['CITY', 'Duration', 'has_children', 'marital_status', 'home_market_value', 'Properties_Market_Value','Vulnerable to theft', 'Potential client']
columns_to_extract_dt2 = ['Annual_prime', 'Insured_Amount','Nombre_Thefts_last5years']
dt = pd.concat([dt1[columns_to_extract_dt1], dt2[columns_to_extract_dt2]], axis=1)
print(dt)
```

```python
class_0 = dt[dt['Potential client'] == 0]
class_1 = dt[dt['Potential client'] == 1]

class_0_sample = class_0.sample(n=7000, random_state=42)
class_1_sample = class_1.sample(n=7000, random_state=42)

dt_balanced = pd.concat([class_0_sample, class_1_sample])
dt = dt_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

print(dt['Potential client'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df= dt[['CITY', 'marital_status']]
```

```python
def encode_df(dataframe):
    le = LabelEncoder()
    for column in dataframe.columns:
        dataframe[column] = le.fit_transform(dataframe[column])
    return dataframe

#encode the dataframe
encode_df(df)
```

```python
del dt['CITY']
```

```python
dt = pd.concat([dt, df], axis=1)
```

```python
numeric_cols = dt.select_dtypes(include=['int64','float64'])

corr_with_target = numeric_cols.corr()['Potential client'].sort_values(ascending=False)

print("CorrÃ©lations avec Potential client:")
print(corr_with_target)

strong_corr = corr_with_target[abs(corr_with_target) > 0.1]
print("\nVariables avec forte corrÃ©lation (>0.1 ou <-0.1):")
print(strong_corr)
```

```python
del dt['marital_status']
```

```python
dt.columns
```

```python
numeric_cols = ['Duration', 'home_market_value', 'Properties_Market_Value',
                'Vulnerable to theft', 'Annual_prime', 'Insured_Amount', 'Nombre_Thefts_last5years']

for col in numeric_cols:
    dt[col+'_norm'] = (dt[col] - dt[col].min()) / (dt[col].max() - dt[col].min())

np.random.seed(42)
dt['fuzzy_score'] = (
    0.2 * dt['Duration_norm'] + 
    0.15 * dt['Properties_Market_Value_norm'] +
    0.15 * dt['Annual_prime_norm'] +
    0.1 * dt['Insured_Amount_norm'] -
    0.2 * dt['Nombre_Thefts_last5years_norm'] -
    0.1 * dt['Vulnerable to theft_norm'] +
    0.05 * dt['has_children'] +
    0.1 * np.random.normal(0, 0.05, dt.shape[0])
)

median_score = dt['fuzzy_score'].median()
dt['Potential_client'] = (dt['fuzzy_score'] > median_score).astype(int)

print(dt['Potential_client'].value_counts())
```

```python
X = dt[['Duration', 'has_children', 'home_market_value', 'Properties_Market_Value',
        'Vulnerable to theft', 'Annual_prime', 'Insured_Amount', 'Nombre_Thefts_last5years']]
y = dt['Potential_client']
```

```python
# Split data 80%-20% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print ('Training cases: %d\nTest cases: %d' % (X_train.shape[0], X_test.shape[0]))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print('Predicted labels: ', predictions)
print('Actual labels:    ', y_test)
print('Accuracy: ', accuracy_score(y_test, predictions))
```

```python
from sklearn. metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import precision_score

print("Overall Precision:", precision_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix

# Print the confusion matrix
cm = confusion_matrix(y_test, predictions)
print (cm)
```

```python
y_scores = model.predict_proba(X_test)
print(y_scores)
```
